```python
import pandas as pd

import pymatchit
```

```python
import pyreadstat
```

```python
import matchingtools as mt
df = mt.load_data(r"C:\Users\balan\OneDrive\Desktop\FGV\Doutorado\Monitoria\STATS1B\Lab 4 - Matching\data\lab04-DATAlab_atualizado.dta")
```

```python
df['gender'] = pd.Categorical(df['gender'])
df['educ'] = pd.Categorical(df['educ'])
df['munsize'] = pd.Categorical(df['munsize'])
df['income'] = pd.Categorical(df['income'])
df['region'] = pd.Categorical(df['region'])
```

```python
mform = "bf ~ age + C(gender) + hdi + C(region) + C(income) + I(hdi ** 2)"
```

```python
fit = pymatchit.matchit(mform,df,'nearest')
```

```python
import matplotlib.pyplot as plt
```

```python
pymatchit.love_plot(fit, threshold=0.1)
```

```python
pymatchit.ecdf_plot(fit, "age")
```

```python
pymatchit.qq_plot(fit, "income")
```
